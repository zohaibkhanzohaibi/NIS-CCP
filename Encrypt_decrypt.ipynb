**Encryption Functions**

```python
def shift_enc(plain_text,k):
  cipher_text=""
  for i in plain_text:
    if i==" ":
      cipher_text += i
      continue
    temp = (ord(i) - ord("A") + (ord(k) - ord("A") + 1)) % 26
    cipher_char = chr(temp + ord("A"))
    cipher_text += cipher_char
  return cipher_text
```

```python
def Vigenere_enc(plain_text, key):
  vig_text=""
  n=len(key)
  i=0
  for ct in plain_text:
    if(i==n):
        i=0
    if(ct==" "):
      vig_text += ct
      continue;
    temp=(ord(ct)-ord("A")+ (ord(key[i])-ord("A")+1))%26
    cipher_char = chr(temp + ord("A"))
    vig_text += cipher_char
    i+=1
  return vig_text
```

```python
def encrypt(plain_text,key):
  plain_copy=plain_text

  for k in key:
    shifted_text = shift_enc(plain_copy, k)
    final_cipher_text = Vigenere_enc(shifted_text, key)
    plain_copy=final_cipher_text
  return plain_copy
```

**Decryption Functions**

```python
def Vigenere_dec(plain_text, key):
  vig_text=""
  n=len(key)
  i=0
  for ct in plain_text:
    if(i==n):
        i=0

    if(ct==" "):
      vig_text += ct
      continue;
    temp=(ord(ct)-ord("A")- (ord(key[i])-ord("A")+1))%26
    if temp<0:
      temp=26+temp
    cipher_char = chr(temp + ord("A"))
    vig_text += cipher_char
    i+=1
  return vig_text
```

```python
def shift_dec(plain_text,k):
  cipher_text=""
  for i in plain_text:
    if i==" ":
      cipher_text += i
      continue
    temp = (ord(i) - ord("A") - (ord(k) - ord("A") + 1)) % 26
    if temp<0:
      temp=26+temp
    cipher_char = chr(temp + ord("A"))
    cipher_text += cipher_char
  return cipher_text
```

```python

def decrypt(cipher_text,key):
  cipher_copy=cipher_text

  for k in key:
    text = Vigenere_dec(cipher_copy, key)
    shifted_text = shift_dec(text, k)
    cipher_copy=shifted_text
  return cipher_copy
```

**Main code**

```python
plain_text="ONE FOR ALL"
Key="AB"
print("Plain text:",plain_text)

cipher_text=encrypt(plain_text,Key)
print("Encrypted Text:", cipher_text)

decrypted_text=decrypt(cipher_text,Key)
print("Decrypted Text:", decrypted_text)
```
